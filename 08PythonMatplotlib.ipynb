```python
from matplotlib.pyplot import *
from numpy import *
%matplotlib inline
#matplotlib.rcParams['figure.facecolor'] = '102020'
```

```python
plot(range(10))
xlabel('measured')
ylabel('calculated')
title('Measured vs. calculated')
grid(True)
#We get a reference to the plot
my_plot = gca()
#and to our line we plotted, which is the first in the plot:
line = my_plot.lines[0]
line.set_marker('o')
setp(line, color='g')
```

```python
x = arange(100)
linear = arange(100)
square = [v * v for v in arange(0,10,0.1)]
lines = plot(x, linear, x, square)
legend(('linear', 'square'))
```

```python
lines = plot(x, linear, 'g:+', x, square, 'r--o')
legend(('linear', 'square'), loc='upper left')
```

# Tick Locators

```python
plot(arange(100))
major_locator = MultipleLocator(10)
minor_locator = MultipleLocator(5)
ax=gca()


ax.xaxis.set_major_locator(major_locator)
ax.xaxis.set_minor_locator(minor_locator)
```

# Contour Plots

```python
x = y = arange(10)
z = ones((10, 10))
z[5,5] = 7
z[2,1] = 3
z[8,7] = 4
contour(x,y,z)
```

```python
v=x
contourf(x,x,z,v)
```
